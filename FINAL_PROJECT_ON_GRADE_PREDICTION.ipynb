```python
import pandas as pd
data=pd.read_excel("grade_horoscope.xlsx")
data.head(5)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import *
from sklearn.metrics import *
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB


train,test=train_test_split(data,test_size=0.3)
print(train.shape)
print(test.shape)
```

```python
train_x=train[["Medu","Fedu","traveltime","studytime","famrel","freetime","alcohol","absences","G1","G2"]]
train_y=train.G3
test_x=test[["Medu","Fedu","traveltime","studytime","famrel","freetime","alcohol","absences","G1","G2"]]
test_y=test.G3
print(train_x.head(5))
print(train_y.head(5))
```

```python
lr= LogisticRegression()
lr.fit(train_x,train_y)
predictions= lr.predict(test_x)
print ("accuracy of svm", accuracy_score(predictions,test_y))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```

```python
from sklearn.metrics import *
from sklearn import svm
model= svm.SVC()
model.fit(train_x,train_y)
prediction= model.predict(test_x)
print ("accuracy of svm", accuracy_score(prediction,test_y))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```

```python
#knn

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=29)
knn.fit(train_x,train_y)
prediction= knn.predict(test_x)
print ("accuracy of knn", accuracy_score(prediction,test_y))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```

```python
#DecisionTreeClassifier
dct=DecisionTreeClassifier(random_state=0)
dct.fit(train_x,train_y)
prediction= dct.predict(test_x)
print(accuracy_score(test_y, prediction))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```

```python
#NaiveBayes
nb=GaussianNB()
nb.fit(train_x,train_y)
predictions= nb.predict(test_x)
print(accuracy_score(test_y, predictions))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```

```python
#LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis()
lda.fit(train_x,train_y)
predictions= lda.predict(test_x)
print(accuracy_score(test_y, predictions))
print("Mean Absoulte Error:", mean_absolute_error(predictions, test_y))
```
